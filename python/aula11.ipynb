```python
from random import randint, random
numeros = [randint(0,20) for _ in range(100)]
dicionario = {}
for numero in numeros:
    if numero not in dicionario:
        dicionario[numero] = 1
    else:
        dicionario[numero] += 1

print(numeros)
print(f'numero: quantidade')
for numero in dicionario:
    print(f'{numero} : {dicionario[numero]}')
```

```python
from random import randint, random

numeros = [randint(2,12) for _ in range(1000)]
dicionario = {}
for numero in numeros:
    if numero not in dicionario:
        dicionario[numero] = 1
    else:
        dicionario[numero] += 1

print(numeros)
print(f'numero: quantidade')
for numero in dicionario:
    print(f'{numero} : {(dicionario[numero])/10:.2f}%')
```
