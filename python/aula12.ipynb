```python
from random import randint

for i in range(100):
    with open ('numeros.txt', 'a') as numeros:
        numeros.write(f'{randint(1,101)},')

def soma_numeros(arquivo):
    with open(arquivo) as arquivo_numeros:
        numeros = arquivo_numeros.readline().strip().split(',')

    numeros.pop()

    for i in range(len(numeros)):
        numeros[i] = int(numeros[i])

    return sum(numeros)

print(soma_numeros('numeros.txt'))
```
